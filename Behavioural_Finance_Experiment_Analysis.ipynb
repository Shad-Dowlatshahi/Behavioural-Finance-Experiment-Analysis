```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Behavioural Finance Experiment - experiment_data.csv")

# Encode risky choice
df["risky_choice"] = df["Choice"].apply(lambda x: 1 if x.lower() == "risky" else 0)

# Calculate overall risk-taking proportion
overall_risk = df["risky_choice"].mean()
print("\nOverall proportion of subjects choosing risky option:", round(overall_risk, 3))

# Framing effect analysis
print("\nProportion of subjects choosing the risky option for each frame:")
print(df.groupby("Frame")["risky_choice"].mean().round(3))

# Confidence level analysis
print("\nProportion of subjects choosing the risky option based on confidence level:")
print(df.groupby("Confidence_Rating")["risky_choice"].mean().round(3))

# Financial literacy analysis
print("\nProportion of subjects choosing the risky option for each level of financial literacy:")
print(df.groupby("Financial_Literacy")["risky_choice"].mean().round(3))

# Bar Chart
df.groupby("Frame")["risky_choice"].mean().plot(kind="bar")
plt.ylabel("Proportion Choosing Risky Option")
plt.title("Risk-Taking by Framing")
plt.show()
```
